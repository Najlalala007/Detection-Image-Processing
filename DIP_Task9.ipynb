```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display, Image
```

```python
# Path to the uploaded file
image_path = 'image.jpg'
display(Image(filename='image.jpg'))
```

```python
# Load the image in grayscale
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
```

```python
# Apply edge detection methods
# 1. Roberts operator approximation using simple convolution kernels
kernel_roberts_x = np.array([[1, 0], [0, -1]])
kernel_roberts_y = np.array([[0, 1], [-1, 0]])
roberts_x = cv2.filter2D(image, -1, kernel_roberts_x)
roberts_y = cv2.filter2D(image, -1, kernel_roberts_y)
roberts = cv2.addWeighted(roberts_x, 0.5, roberts_y, 0.5, 0)
```

```python
# 2. Prewitt operator
kernel_prewitt_x = np.array([[1, 0, -1], [1, 0, -1], [1, 0, -1]])
kernel_prewitt_y = np.array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]])
prewitt_x = cv2.filter2D(image, -1, kernel_prewitt_x)
prewitt_y = cv2.filter2D(image, -1, kernel_prewitt_y)
prewitt = cv2.addWeighted(prewitt_x, 0.5, prewitt_y, 0.5, 0)
```

```python
# 3. Sobel operator
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)
sobel = cv2.addWeighted(cv2.convertScaleAbs(sobel_x), 0.5, cv2.convertScaleAbs(sobel_y), 0.5, 0)
```

```python
# 4. Canny edge detection
canny = cv2.Canny(image, 100, 200)
```

```python
# Display the results
plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis('off')

plt.subplot(2, 3, 2)
plt.imshow(roberts, cmap='gray')
plt.title("Roberts Edge Detection")
plt.axis('off')

plt.subplot(2, 3, 3)
plt.imshow(prewitt, cmap='gray')
plt.title("Prewitt Edge Detection")
plt.axis('off')

plt.subplot(2, 3, 4)
plt.imshow(sobel, cmap='gray')
plt.title("Sobel Edge Detection")
plt.axis('off')

plt.subplot(2, 3, 5)
plt.imshow(canny, cmap='gray')
plt.title("Canny Edge Detection")
plt.axis('off')

plt.tight_layout()
plt.show()
```
